# Prueba técnica P1

Combine los 3 archivos bikes.xlsx (Tipos de Bicicleta), orderlines.xlsx (Ordenes de pedidos), bikeshops.xlsx (Tiendas de bicicleta) en un solo archivo. Divida los nuevos datos de la tabla en 2 categorías de montaña y carretera utilizando la columna de descripción para obtener los ingresos totales para bicicletas de montaña y de carretera


```python
df_merge1=pd.merge(df_orderlines,df_bikes,how='inner',left_on='product.id',right_on='bike.id')
df_merged=pd.merge(df_merge1,df_bikeshops,how='inner',left_on='order.line',right_on='bikeshop.id')
df_merged.drop(columns=['product.id','order.line'],inplace=True)
```

```python
df_merged['Category']=df_merged['description'].map(lambda x: x.split('-')[0].strip())
```

```python
df_merged.head()
```

Para obtener los ingresos totales, utiliza la siguiente ecuación:

> Revenue = $quantity * price$ <br>
> Mountain: \\$ xx,xxx,xxx  <br>
> Road: \\$ xx,xxx,xxx  <br>

```python
df_merged['Revenue']=df_merged['quantity']*df_merged['price']
```

```python
df_merged.head()
```

```python
result = df_merged.groupby('Category')['Revenue'].sum()
result_P1 = []
for category, revenue in result.items():
    result_P1.append(f"{category}: $ {revenue:,}")

result_P1
```

```python
print("\n".join(result_P1))
```

# Prueba técnica P2

Utiliza las columnas de model y location para crear una nueva columna States con los estados que están vendiendo solo el modelo Supersix

```python
df_merged['states'] = df_merged['location'].map(lambda x: x.split(',')[1].strip())
```

```python
df_merged.head()
```

```python
mask = df_merged['model'].str.contains('Supersix')
df_filtered = df_merged[mask]
result_P2=df_filtered[['order.date','model','Category','quantity','price','location','Revenue','states']].reset_index(drop=True)
result_P2.head(10)
```

# Prueba técnica P3

Crea una nueva columna que contenga los cuantiles altos, medios y bajos de las ventas (Revenue) del modelo Supersix en el estado de Florida (FL) <br>
High    > 75%<br>
Medium  >= 25% & <= 75%<br>
Low     < 25%

```python
result_P3 = result_P2[result_P2['states'] == 'FL'].copy().reset_index(drop=True).reset_index(drop=True)
quantiles = result_P3['Revenue'].quantile([0.25, 0.75])

def Supersix_quantile(revenue):
    if revenue > quantiles[0.75]:
        return 'High'
    elif revenue >= quantiles[0.25] and revenue <= quantiles[0.75]:
        return 'Medium'
    else:
        return 'Low'

result_P3['Quantile'] = result_P3['Revenue'].apply(Supersix_quantile)

result_P3
```
